```python
class Cliente:
  def __init__(self, id, nombre, telefono, correo, ciudad, estado):
    self.id = id
    self.nombre = nombre
    self.telefono = telefono
    self.correo = correo
    self.ciudad = ciudad
    self.estado = estado

  def mostrar_info(self):
      print(f"Id: {self.id}, Nombre: {self.nombre}, Teléfono: {self.telefono}, Correo: {self.correo}, Ciudad: {self.ciudad}, Estado: {self.estado.strip()}")

  def mostrar_menu():
    print("Menú Principal")
    print("Opción 1: Registrar cliente")
    print("Opción 2: Listar clientes")
    print("Opción 3: Buscar cliente por su número de identificación y mostrar sus datos")
    print("Opción 4: Buscar cliente por nombre completo y mostrar sus datos.")
    print("Opción 5: Consultar clientes por ciudad y listar sus datos.")
    print("Opción 6: Mostrar solo clientes activos.")
    print("Opción 7: Mostrar solo clientes inactivos")
    print("Opción 8: Consultar cliente por correo electrónico y mostrar todos sus datos.")
    print("Opción 9: Mostrar clientes cuyo nombre empiece por una letra específica.")
    print("Opción 10: Mostrar clientes cuyo nombre contenga una palabra específica.")
    print("Opción 11: Consultar cliente por número telefónico.")
    print("Opción 12: Mostrar total de clientes registrados.")
    print("Opción 13: Mostrar total de clientes activos.")
    print("Opción 14: Mostrar total de clientes inactivos.")
    print("Opción 15: Contar clientes por ciudad.")
    print("Opción 16: Mostrar la ciudad con mayor número de clientes.")
    print("Opción 17: Mostrar la ciudad con menor número de clientes.")
    print("Opción 18: Calcular porcentaje de clientes activos.")
    print("Opción 19: Calcular porcentaje de clientes inactivos.")
    print("Opción 20: Mostrar clientes activos de una ciudad específica.")
    print("Opción 21: Mostrar clientes inactivos de una ciudad específica.")
    print("Opción 22: Mostrar clientes dentro de un rango de identificaciones.")
    print("Opción 23: Mostrar clientes con nombres duplicados.")
    print("Opción 24: Mostrar clientes ordenados por estado del cliente.")
    print("Opción 25: Mostrar clientes ordenados por número de identificación.")
```

```python
class Archivo:
  def __init__(self, ruta):
    self.ruta = ruta
    self.clientes = []

  def crear_coleccion(self):
    self.clientes = []
    with open(self.ruta, "r") as archivo:
      for linea in archivo:
        id, nombre, telefono, correo, ciudad, estado = linea.strip().split(";")
        cliente = Cliente(id, nombre, telefono, correo, ciudad, estado)
        self.clientes.append(cliente)

  def registrar_clientes(self):
    id_nuevo = input("Ingrese el número de identificación: ")

    # Validate if ID already exists
    for cliente in self.clientes:
      if cliente.id == id_nuevo:
        print(f"Error: Ya existe un cliente con la identificación {id_nuevo}. No se puede registrar.")
        return

    nombre = input("Ingrese el nombre completo: ")
    telefono = input("Ingrese el número de teléfono: ")
    correo = input("Ingrese el correo electrónico: ")
    ciudad = input("Ingrese la ciudad: ")
    estado = input("Ingrese el estado del cliente (Activo/Inactivo): ")
    cliente = Cliente(id_nuevo, nombre, telefono, correo, ciudad, estado)
    self.clientes.append(cliente)
    with open(self.ruta, "a") as archivo:
      archivo.write(f"{id_nuevo};{nombre};{telefono};{correo};{ciudad};{estado}\n")
    print(f"Cliente con ID {id_nuevo} registrado exitosamente.")

  def listar_clientes(self):
    for cliente in self.clientes:
      cliente.mostrar_info()

  def buscar_cliente_id(self):
    id = input("Ingrese el número de identificación del cliente: ")
    for cliente in self.clientes:
      if cliente.id == id:
        cliente.mostrar_info()
        return

  def buscar_cliente_nombre(self):
    nombre = input("Ingrese el nombre completo del cliente: ")
    for cliente in self.clientes:
      if cliente.nombre == nombre:
        cliente.mostrar_info()
        return

  def consultar_clientes_ciudad(self):
    ciudad = input("Ingrese la ciudad: ")
    for cliente in self.clientes:
      if cliente.ciudad == ciudad:
        cliente.mostrar_info()

  def mostrar_clientes_activos(self):
    for cliente in self.clientes:
      if cliente.estado == "Activo":
        cliente.mostrar_info()

  def mostrar_clientes_inactivos(self):
    for cliente in self.clientes:
      if cliente.estado == "Inactivo":
        cliente.mostrar_info()

  def consultar_cliente_correo(self):
    correo = input("Ingrese el correo electrónico del cliente: ")
    for cliente in self.clientes:
      if cliente.correo == correo:
        cliente.mostrar_info()
        return

  def mostrar_clientes_letra(self):
    letra = input("Ingrese la letra: ")
    for cliente in self.clientes:
      if cliente.nombre.startswith(letra):
        cliente.mostrar_info()

  def mostrar_clientes_palabra(self):
    palabra = input("Ingrese la palabra: ")
    for cliente in self.clientes:
      if palabra in cliente.nombre:
        cliente.mostrar_info()

  def consultar_cliente_telefono(self):
    telefono = input("Ingrese el número de teléfono del cliente: ")
    for cliente in self.clientes:
      if cliente.telefono == telefono:
        cliente.mostrar_info()
        return

  def mostrar_total_clientes(self):
    print(f"Total de clientes registrados: {len(self.clientes)}")

  def mostrar_total_clientes_activos(self):
    activos = sum(1 for cliente in self.clientes if cliente.estado == "Activo")
    print(f"Total de clientes activos: {activos}")

  def mostrar_total_clientes_inactivos(self):
    inactivos = sum(1 for cliente in self.clientes if cliente.estado == "Inactivo")
    print(f"Total de clientes inactivos: {inactivos}")

  def contar_clientes_ciudad(self):
    ciudad = input("Ingrese la ciudad: ")
    contador = sum(1 for cliente in self.clientes if cliente.ciudad == ciudad)
    print(f"Total de clientes en la ciudad {ciudad}: {contador}")

  def mostrar_ciudad_mayor(self):
    ciudades = {}
    for cliente in self.clientes:
      if cliente.ciudad in ciudades:
        ciudades[cliente.ciudad] += 1
      else:
        ciudades[cliente.ciudad] = 1
    ciudad_mayor = max(ciudades, key=ciudades.get)
    print(f"La ciudad con mayor número de clientes es: {ciudad_mayor}")

  def mostrar_ciudad_menor(self):
    ciudades = {}
    for cliente in self.clientes:
      if cliente.ciudad in ciudades:
        ciudades[cliente.ciudad] += 1
      else:
        ciudades[cliente.ciudad] = 1
    ciudad_menor = min(ciudades, key=ciudades.get)
    print(f"La ciudad con menor número de clientes es: {ciudad_menor}")

  def mostrar_porcentaje_activos(self):
    activos = sum(1 for cliente in self.clientes if cliente.estado == "Activo")
    inactivos = sum(1 for cliente in self.clientes if cliente.estado == "Inactivo")
    total = activos + inactivos
    porcentaje_activos = (activos / total) * 100
    print(f"El porcentaje de clientes activos es: {porcentaje_activos:.2f}%")

  def mostrar_porcentaje_inactivos(self):
    activos = sum(1 for cliente in self.clientes if cliente.estado == "Activo")
    inactivos = sum(1 for cliente in self.clientes if cliente.estado == "Inactivo")
    total = activos + inactivos
    porcentaje_inactivos = (inactivos / total) * 100
    print(f"El porcentaje de clientes inactivos es: {porcentaje_inactivos:.2f}%")

  def mostrar_clientes_ciudad_activa(self):
    ciudad = input("Ingrese la ciudad: ")
    for cliente in self.clientes:
      if cliente.estado == "Activo" and cliente.ciudad == ciudad:
        cliente.mostrar_info()

  def mostrar_clientes_ciudad_inactiva(self):
    ciudad = input("Ingrese la ciudad: ")
    for cliente in self.clientes:
      if cliente.estado == "Inactivo" and cliente.ciudad == ciudad:
        cliente.mostrar_info()

  def mostrar_clientes_rango_id(self):
    id_inicio = input("Ingrese el número de identificación inicial: ")
    id_fin = input("Ingrese el número de identificación: ")
    for cliente in self.clientes:
      if id_inicio <= cliente.id <= id_fin:
        cliente.mostrar_info()

  def mostrar_clientes_duplicados(self):
    nombres = {}
    for cliente in self.clientes:
      if cliente.nombre in nombres:
        nombres[cliente.nombre] += 1
      else:
        nombres[cliente.nombre] = 1

    for nombre, contador in nombres.items():
      if contador > 1:
        print(f"El nombre {nombre} se repite {contador} veces.")
    else:
      print("No hay clientes con nombres duplicados.")

  def mostrar_clientes_estado(self):
    for cliente in sorted(self.clientes, key=lambda cliente: cliente.estado):
      cliente.mostrar_info()

  def mostrar_clientes_id(self):
    for cliente in sorted(self.clientes, key=lambda cliente: cliente.id):
      cliente.mostrar_info()

  def modificar_cliente(self):
    id_a_modificar = input("Ingrese el número de identificación del cliente a modificar: ")
    cliente_encontrado = None
    for cliente in self.clientes:
      if cliente.id == id_a_modificar:
        cliente_encontrado = cliente
        break

    if cliente_encontrado:
      print(f"Modificando cliente con ID: {id_a_modificar}")
      cliente_encontrado.nombre = input("Ingrese el nuevo nombre: ")
      cliente_encontrado.telefono = input("Ingrese el nuevo número de teléfono: ")
      cliente_encontrado.correo = input("Ingrese el nuevo correo electrónico: ")
      cliente_encontrado.ciudad = input("Ingrese la nueva ciudad: ")
      cliente_encontrado.estado = input("Ingrese el nuevo estado del cliente (Activo/Inactivo): ")
      print("Cliente modificado en memoria.")

      with open(self.ruta, "w") as archivo:
        for cliente_actual in self.clientes:
          archivo.write(f"{cliente_actual.id};{cliente_actual.nombre};{cliente_actual.telefono};{cliente_actual.correo};{cliente_actual.ciudad};{cliente_actual.estado}\n")
      print("Archivo actualizado con el cliente modificado.")
    else:
      print(f"No se encontró ningún cliente con ID: {id_a_modificar}")
```

```python
def ejecutar_opcion(opcion, archiv):
  if opcion == 1:
    archiv.registrar_clientes()
  elif opcion == 2:
    archiv.listar_clientes()
  elif opcion == 3:
    archiv.buscar_cliente_id()
  elif opcion == 4:
    archiv.buscar_cliente_nombre()
  elif opcion == 5:
    archiv.consultar_clientes_ciudad()
  elif opcion == 6:
    archiv.mostrar_clientes_activos()
  elif opcion == 7:
    archiv.mostrar_clientes_inactivos()
  elif opcion == 8:
    archiv.consultar_cliente_correo()
  elif opcion == 9:
    archiv.mostrar_clientes_letra()
  elif opcion == 10:
    archiv.mostrar_clientes_palabra()
  elif opcion == 11:
    archiv.consultar_cliente_telefono()
  elif opcion == 12:
    archiv.mostrar_total_clientes()
  elif opcion == 13:
    archiv.mostrar_total_clientes_activos()
  elif opcion == 14:
    archiv.mostrar_total_clientes_inactivos()
  elif opcion == 15:
    archiv.contar_clientes_ciudad()
  elif opcion == 16:
    archiv.mostrar_ciudad_mayor()
  elif opcion == 17:
    archiv.mostrar_ciudad_menor()
  elif opcion == 18:
    archiv.mostrar_porcentaje_activos()
  elif opcion == 19:
    archiv.mostrar_porcentaje_inactivos()
  elif opcion == 20:
    archiv.mostrar_clientes_ciudad_activa()
  elif opcion == 21:
    archiv.mostrar_clientes_ciudad_inactiva()
  elif opcion == 22:
    archiv.mostrar_clientes_rango_id()
  elif opcion == 23:
    archiv.mostrar_clientes_duplicados()
  elif opcion == 24:
    archiv.mostrar_clientes_estado()
  elif opcion == 25:
    archiv.mostrar_clientes_id()
```
